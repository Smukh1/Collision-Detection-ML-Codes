```python
import numpy as np
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
# Generate a simple time series
time_steps = 10
data = np.array([i for i in range(time_steps)])

# Create input and output sequences
X = []
y = []
for i in range(len(data) - 1):
    X.append(data[i:i+1])
    y.append(data[i+1])

print(data)
```

```python
# Convert to numpy arrays
X = np.array(X)
y = np.array(y)

# Reshape input data for LSTM
X = np.reshape(X, (X.shape[0], X.shape[1], 1)) # expected format for LSTM: (samples, time_steps, features)
```

```python
# Build the LSTM model (an LSTM layer and a Dense output layer)
model = Sequential()
model.add(LSTM(units=50, input_shape=(1, 1)))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X, y, epochs=100, batch_size=1, verbose=2)
```

```python
# Make a prediction
test_data = np.array([9])
test_data = np.reshape(test_data, (1, 1, 1))
prediction = model.predict(test_data)
print(f"Predicted next value: {prediction[0][0]}")
```
